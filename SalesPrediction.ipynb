```python
df['budget_gap'] = df['budget'] - df['car_price']
df['followup_efficiency'] = df['followup_count'] / (df['followup_delay_days']+1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["loss_reason_encoded"] = le.fit_transform(df["loss_reason"])
```

```python
x = df.drop(columns=["customer_id", "sales_executive_id", "will_buy","budget", "car_price", 'followup_count', 'followup_delay_days','loss_reason','loss_reason_encoded'])
y = df['will_buy']
z = df['loss_reason_encoded']
```

```python
x.columns
```

```python

binary_cols = [
    "test_drive_taken",
   
    "competitor_visit"
]
for col in binary_cols:
    x[col] = x[col].astype(int)


x["gender"] = x["gender"].map({"Male": 1, "Female": 0})
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe_cols = ["city", "brand", "car_model", "fuel_type"]
ohe = OneHotEncoder(
    sparse_output=False,
    drop = "first",
    handle_unknown="ignore"
)
ohe.fit(x[ohe_cols])

ohe_array = ohe.transform(x[ohe_cols])

ohe_feature_names = ohe.get_feature_names_out(ohe_cols)
ohe_df = pd.DataFrame(ohe_array, columns=ohe_feature_names, index=x.index)



```

```python

np.random.seed(42)
flip_prob = 0.15  

if "test_drive_taken" in x.columns:
    original_sum = x["test_drive_taken"].sum()
    
    flip_mask = np.random.random(len(x)) < flip_prob
    x.loc[flip_mask, "test_drive_taken"] = 1 - x.loc[flip_mask, "test_drive_taken"]
    
    new_sum = x["test_drive_taken"].sum()
    flipped_count = flip_mask.sum()
    
    print(f"Flipped {flipped_count} values ({flip_prob*100:.0f}%) in test_drive_taken")
    print(f"Before: {int(original_sum)} ones, After: {int(new_sum)} ones")
else:
    print("test_drive_taken not in features")
```

```python
x['test_drive_taken'].unique()
```

```python
x = x.drop(columns=ohe_cols)
x = pd.concat([x, ohe_df], axis=1)
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test, z_train, z_test = train_test_split(
    x, y, z, test_size=0.2, random_state=42,stratify=y
)
```

```python
from imblearn.over_sampling import SMOTE

print("Original class distribution:")
print(y_train.value_counts())

smote = SMOTE(random_state=42, k_neighbors=3)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

print("\nAfter SMOTE:")
print(pd.Series(y_train_balanced).value_counts())
```

```python
'''import category_encoders as ce

te_cols = ["city", "brand", "car_model", "previous_brand_owned", "fuel_type", "trust_index"]

encoder = ce.TargetEncoder(cols=te_cols)

X_train = encoder.fit_transform(X_train, y_train)

X_test = encoder.transform(X_test)'''
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# Balanced complexity - not too simple, not too complex
model = RandomForestClassifier(
    n_estimators=150,
    max_depth=4,
    min_samples_split=15,
    min_samples_leaf=8,
    max_features='sqrt',
    class_weight="balanced", 
    random_state=42
)
model.fit(X_train_balanced, y_train_balanced)
print('RandomForest - Training Score: ',model.score(X_train_balanced, y_train_balanced))
print('RandomForest - Testing Score: ',round(model.score(X_test, y_test),5))

model1 = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.02,
    max_depth=4,
    min_samples_split=15,
    min_samples_leaf=8,
    subsample=0.8,
    random_state=42
)
model1.fit(X_train_balanced, y_train_balanced)
print('\nGradientBoosting - Training Score: ',model1.score(X_train_balanced, y_train_balanced))
print('GradientBoosting - Testing Score: ',round(model1.score(X_test, y_test),5))

# XGBoost with moderate regularization (not too aggressive)
model2 = XGBClassifier(
    n_estimators=300,
    max_depth=3,
    learning_rate=0.02,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=3,
    gamma=0.5,
    reg_alpha=0.5,
    reg_lambda=1.0,
    scale_pos_weight=1.0,
    eval_metric="logloss",
    random_state=42
)
model2.fit(X_train_balanced, y_train_balanced) 
print("\nXGBoost Purchase Model - Train:", model2.score(X_train_balanced, y_train_balanced))
print("XGBoost Purchase Model - Test :", model2.score(X_test, y_test))

# Reason model (only for non-buyers) with moderate regularization
reason_mask_train = y_train == 0
X_reason_train = X_train[reason_mask_train]
z_reason_train = z_train[reason_mask_train]

reason_mask_test = y_test == 0
X_reason_test = X_test[reason_mask_test]
z_reason_test = z_test[reason_mask_test]

if "No Reason" in le.classes_:
    no_reason_label = le.transform(["No Reason"])[0]
    train_mask = z_reason_train != no_reason_label
    test_mask = z_reason_test != no_reason_label
    X_reason_train_filtered = X_reason_train[train_mask]
    z_reason_train_filtered = z_reason_train[train_mask]
    X_reason_test_filtered = X_reason_test[test_mask]
    z_reason_test_filtered = z_reason_test[test_mask]
else:
    X_reason_train_filtered = X_reason_train
    z_reason_train_filtered = z_reason_train
    X_reason_test_filtered = X_reason_test
    z_reason_test_filtered = z_reason_test

# Apply SMOTE to balance reason classes
smote_reason = SMOTE(random_state=42, k_neighbors=2)
X_reason_balanced, z_reason_balanced = smote_reason.fit_resample(X_reason_train_filtered, z_reason_train_filtered)

# Train XGBoost reason model with moderate regularization
num_reason_classes = len(np.unique(z_reason_balanced))
reason_model = XGBClassifier(
    n_estimators=200,
    max_depth=3,
    learning_rate=0.02,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=3,
    gamma=0.5,
    reg_alpha=0.5,
    reg_lambda=1.0,
    objective="multi:softprob",
    num_class=num_reason_classes,
    eval_metric="mlogloss",
    random_state=42
)
reason_model.fit(X_reason_balanced, z_reason_balanced)

print("\nReason Model (XGBoost) - Train:", reason_model.score(X_reason_balanced, z_reason_balanced))
print("Reason Model (XGBoost) - Test :", reason_model.score(X_reason_test_filtered, z_reason_test_filtered))
```

```python
from sklearn.metrics import (accuracy_score, classification_report, 
                             confusion_matrix, f1_score, recall_score, 
                             precision_score, roc_auc_score)
import seaborn as sns


print("RANDOM FOREST EVALUATION....")

y_pred = model.predict(X_test)
y_pred_proba = model.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall (Sensitivity): {recall_score(y_test, y_pred):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred):.4f}")
print(f"ROC-AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print()
print("GRADIENT BOOSTING EVALUATION....")

y_pred = model1.predict(X_test)
y_pred_proba = model1.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall (Sensitivity): {recall_score(y_test, y_pred):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred):.4f}")
print(f"ROC-AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print()
print("XGBOOST EVALUATION....")

y_pred = model2.predict(X_test)
y_pred_proba = model2.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall (Sensitivity): {recall_score(y_test, y_pred):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred):.4f}")
print(f"ROC-AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))


print()
print("REASON MODEL EVALUATION....")

z_pred = reason_model.predict(X_reason_test_filtered)
z_pred_proba = reason_model.predict_proba(X_reason_test_filtered)

print(f"Accuracy: {accuracy_score(z_reason_test_filtered, z_pred):.4f}")
print(f"Precision: {precision_score(z_reason_test_filtered, z_pred, average='weighted'):.4f}")
print(f"Recall: {recall_score(z_reason_test_filtered, z_pred, average='weighted'):.4f}")
print(f"F1-Score: {f1_score(z_reason_test_filtered, z_pred, average='weighted'):.4f}")

```
